## Data Ingestion

```python
-- For STATE
INSERT INTO STATE (state_id, state_name)
SELECT DISTINCT DENSE_RANK() over (order by food_service_facility_state), food_service_facility_state 
  FROM ####.DATA_RESTAURANT_LAST_INSPECTION 
 WHERE food_service_facility_state IS NOT NULL
;

-- For LOCAL_HEALTH_DEPARTMENT
INSERT INTO LOCAL_HEALTH_DEPARTMENT (LOCAL_HEALTH_DEPARTMENT_id, LOCAL_HEALTH_DEPARTMENT_name)
SELECT DISTINCT DENSE_RANK() over (order by LOCAL_HEALTH_DEPARTMENT), LOCAL_HEALTH_DEPARTMENT 
  FROM ####.DATA_RESTAURANT_LAST_INSPECTION 
 WHERE LOCAL_HEALTH_DEPARTMENT IS NOT NULL
;

-- For INSPECTION_TYPE
INSERT INTO INSPECTION_TYPE (INSPECTION_TYPE_id, INSPECTION_TYPE_description)
SELECT DISTINCT DENSE_RANK() over (order by INSPECTION_TYPE), INSPECTION_TYPE 
  FROM ####.DATA_RESTAURANT_LAST_INSPECTION 
 WHERE INSPECTION_TYPE IS NOT NULL
;

-- For COUNTY
INSERT INTO COUNTY (COUNTY_ID, COUNTY_NAME, STATE_ID)
SELECT DISTINCT DENSE_RANK() over (order by s.food_service_facility_state,s.COUNTY), s.COUNTY, t.STATE_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION s
JOIN STATE t ON t.STATE_NAME = s.food_service_facility_state
ORDER BY 1
;

-- For CITY
INSERT INTO CITY (CITY_ID, CITY_NAME, STATE_ID)
SELECT DISTINCT DENSE_RANK() over (order by s.food_service_facility_state,s.CITY), s.CITY, t.STATE_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION s
JOIN STATE t ON t.STATE_NAME = s.food_service_facility_state
ORDER BY 1
;

-- For MUNICIPALITY
INSERT INTO MUNICIPALITY (MUNICIPALITY_ID, MUNICIPALITY_NAME, STATE_ID)
SELECT DISTINCT DENSE_RANK() over (order by s.food_service_facility_state,s.MUNICIPALITY), s.MUNICIPALITY, t.STATE_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION s
JOIN STATE t ON t.STATE_NAME = s.food_service_facility_state
ORDER BY 1
;

-- For FACILITY
INSERT INTO FACILITY (FACILITY_ID, FACILITY_NAME, DESCRIPTION, NYSDOH_GAZETTEER_1980)
SELECT DISTINCT DENSE_RANK() over (order by FACILITY, DESCRIPTION, NYSDOH_GAZETTEER_1980), FACILITY, DESCRIPTION, NYSDOH_GAZETTEER_1980
FROM ####.DATA_RESTAURANT_LAST_INSPECTION
WHERE FACILITY IS NOT NULL
;

-- For PERM_OPERATOR
INSERT INTO PERM_OPERATOR (PERM_OPERATOR_ID, PERM_OPERATOR_FIRST_NAME, PERM_OPERATOR_LAST_NAME)
SELECT DISTINCT DENSE_RANK() over (order by PERM_OPERATOR_FIRST_NAME, PERM_OPERATOR_LAST_NAME),
PERM_OPERATOR_FIRST_NAME, PERM_OPERATOR_LAST_NAME
FROM ####.DATA_RESTAURANT_LAST_INSPECTION
;

-- For PERMIT
INSERT INTO PERMIT (PERMIT_ID, PERMITTED_D_B_A, PERMITTED_CORP_NAME, PERMIT_EXPIRATION_DATE, FACILITY_ID)
SELECT DISTINCT DENSE_RANK() over (order by d.PERMITTED_D_B_A, d.PERMITTED_CORP_NAME, d.PERMIT_EXPIRATION_DATE, f.FACILITY_ID),
d.PERMITTED_D_B_A, d.PERMITTED_CORP_NAME, d.PERMIT_EXPIRATION_DATE, f.FACILITY_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
 JOIN FACILITY f ON f.facility_name = d.facility
  AND f.description = d.description
  AND f.NYSDOH_GAZETTEER_1980 = d.NYSDOH_GAZETTEER_1980
;

-- For FACILITY_LOCATION
INSERT INTO FACILITY_LOCATION (FACILITY_LOCATION_ID, FACILITY_ADDRESS, ZIP_CODE, LATITUDE, LONGITUDE, MUNICIPALITY_ID, CITY_ID, COUNTY_ID, FOOD_SERVICE_FACILITY_STATE_ID, FACILITY_ID)
SELECT DISTINCT DENSE_RANK() over (order by d.FACILITY_ADDRESS, d.ZIP_CODE, d.location_lat_long, m.MUNICIPALITY_ID, ci.CITY_ID, co.COUNTY_ID, co.state_id, f.facility_id) RANK,
       d.FACILITY_ADDRESS,
       CASE WHEN REGEXP_LIKE(REPLACE(d.ZIP_CODE,' ',''),'^[0-9]+$')
        THEN TO_NUMBER(REPLACE(d.ZIP_CODE,' ',''))
        ELSE NULL END
       , TO_NUMBER(SUBSTR(d.location_lat_long, 2, INSTR(d.location_lat_long, ', ') - 2)),
       TO_NUMBER(SUBSTR(d.location_lat_long, INSTR(d.location_lat_long, ', ') + 1, LENGTH(d.location_lat_long) - INSTR(d.location_lat_long, ',') - 2)),
       m.MUNICIPALITY_ID, ci.CITY_ID, co.COUNTY_ID, co.state_id, f.facility_id
 FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
 JOIN MUNICIPALITY m ON m.MUNICIPALITY_name = d.MUNICIPALITY
 JOIN CITY ci ON ci.CITY_name = d.CITY
 JOIN COUNTY co ON co.COUNTY_name = d.COUNTY
 JOIN FACILITY f ON f.facility_name = d.facility
  AND f.description = d.description
  AND f.NYSDOH_GAZETTEER_1980 = d.NYSDOH_GAZETTEER_1980
;

-- For PERM_OPERATOR_PERMIT
INSERT INTO PERM_OPERATOR_PERMIT (PERM_OPERATOR_ID, PERMIT_ID)
SELECT DISTINCT
       po.PERM_OPERATOR_ID,
       p.PERMIT_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
JOIN PERMIT p ON p.PERMITTED_D_B_A = d.PERMITTED_D_B_A
 AND p.PERMITTED_CORP_NAME = d.PERMITTED_CORP_NAME
 AND p.PERMIT_EXPIRATION_DATE = d.PERMIT_EXPIRATION_DATE
JOIN PERM_OPERATOR po ON po.PERM_OPERATOR_FIRST_NAME = d.PERM_OPERATOR_FIRST_NAME 
 AND po.PERM_OPERATOR_LAST_NAME = d.PERM_OPERATOR_LAST_NAME
;

-- For INSPECTION
INSERT INTO INSPECTION (INSPECTION_ID, INSPECTION_DATE, TOTAL_CRITICAL_VIOLATIONS,
                        TOTAL_CRITICAL_NOT_CORRECTED, TOTAL_NONCRITICAL_VIOLATIONS, INSPECTION_COMMENTS,
                        INSPECTION_TYPE_ID, FACILITY_ID, OPERATION_NAME, NYS_HEALTH_OPERATION_ID)
SELECT DISTINCT DENSE_RANK() over (order by d.LAST_INSPECTED, d.TOTAL_CRITICAL_VIOLATIONS,
    d.TOTAL_CRITICAL_NOT_CORRECTED, d.TOTAL_NONCRITICAL_VIOLATIONS, d.INSPECTION_COMMENTS,
    i.INSPECTION_TYPE_ID, f.FACILITY_ID, d.OPERATION_NAME, d.NYS_HEALTH_OPERATION_ID),
    d.LAST_INSPECTED, d.TOTAL_CRITICAL_VIOLATIONS, d.TOTAL_CRITICAL_NOT_CORRECTED,
    d.TOTAL_NONCRITICAL_VIOLATIONS, d.INSPECTION_COMMENTS, i.INSPECTION_TYPE_ID,
    f.FACILITY_ID, d.OPERATION_NAME, d.NYS_HEALTH_OPERATION_ID
 FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
 JOIN INSPECTION_TYPE i ON i.INSPECTION_TYPE_DESCRIPTION = d.INSPECTION_TYPE
 JOIN FACILITY f ON f.facility_name = d.facility
  AND f.description = d.description
  AND f.NYSDOH_GAZETTEER_1980 = d.NYSDOH_GAZETTEER_1980
;

-- For HEALTH_DEPARTMENT_INSPECTION
INSERT INTO HEALTH_DEPARTMENT_INSPECTION (LOCAL_HEALTH_DEPARTMENT_ID, INSPECTION_ID)
SELECT DISTINCT
       h.LOCAL_HEALTH_DEPARTMENT_ID,
       i.INSPECTION_ID
FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
JOIN INSPECTION i ON i.INSPECTION_DATE = d.LAST_INSPECTED
 AND i.TOTAL_CRITICAL_VIOLATIONS = d.TOTAL_CRITICAL_VIOLATIONS
 AND i.TOTAL_CRITICAL_NOT_CORRECTED = d.TOTAL_CRITICAL_NOT_CORRECTED
 AND i.TOTAL_NONCRITICAL_VIOLATIONS = d.TOTAL_NONCRITICAL_VIOLATIONS
 AND i.INSPECTION_COMMENTS = d.INSPECTION_COMMENTS
 AND i.OPERATION_NAME = d.OPERATION_NAME
 AND i.NYS_HEALTH_OPERATION_ID = d.NYS_HEALTH_OPERATION_ID
JOIN LOCAL_HEALTH_DEPARTMENT h ON h.LOCAL_HEALTH_DEPARTMENT_NAME = d.LOCAL_HEALTH_DEPARTMENT 
;

-- For VIOLATION
DROP TABLE CURR_VIOLATION;
CREATE TABLE CURR_VIOLATION (
    INSPECTION_ID integer NOT NULL,
    VIOLATION_DESCRIPTION varchar2(3000)  NULL
);

INSERT INTO CURR_VIOLATION (INSPECTION_ID, VIOLATION_DESCRIPTION)
SELECT DISTINCT i.inspection_id, d.violations
FROM ####.DATA_RESTAURANT_LAST_INSPECTION d
JOIN INSPECTION i ON i.INSPECTION_DATE = d.LAST_INSPECTED
 AND i.TOTAL_CRITICAL_VIOLATIONS = d.TOTAL_CRITICAL_VIOLATIONS
 AND i.TOTAL_CRITICAL_NOT_CORRECTED = d.TOTAL_CRITICAL_NOT_CORRECTED
 AND i.TOTAL_NONCRITICAL_VIOLATIONS = d.TOTAL_NONCRITICAL_VIOLATIONS
 AND i.INSPECTION_COMMENTS = d.INSPECTION_COMMENTS
 AND i.OPERATION_NAME = d.OPERATION_NAME
 AND i.NYS_HEALTH_OPERATION_ID = d.NYS_HEALTH_OPERATION_ID
WHERE d.violations <> 'No violations found.'
;

DROP TABLE TEST_VIOLATION;
CREATE TABLE TEST_VIOLATION (
    INSPECTION_ID integer NOT NULL,
    VIOLATION_ID varchar2(5)  NULL,
    VIOLATION_DESCRIPTION varchar2(300)  NULL)
;

-- ITERATE THESE 4 BLOCKS OF CODE 22 TIMES UNTIL IT'S ALL [0 ROW(S) UPDATED]
-- START OF ITERATIVE BLOCK
INSERT INTO TEST_VIOLATION (INSPECTION_ID, VIOLATION_ID, VIOLATION_DESCRIPTION)
SELECT INSPECTION_ID,
 REPLACE(SUBSTR(VIOLATION_DESCRIPTION,
       INSTR(VIOLATION_DESCRIPTION, 'Item ') + 5,
       INSTR(VIOLATION_DESCRIPTION, '-') - (INSTR(VIOLATION_DESCRIPTION, 'Item ') + 5)
 ),' ','') AS violation_id
 , SUBSTR(VIOLATION_DESCRIPTION,
       INSTR(VIOLATION_DESCRIPTION, '-') + 2,
       INSTR(VIOLATION_DESCRIPTION, ';') - INSTR(VIOLATION_DESCRIPTION, '-') - 2
 ) AS violation_description
 FROM CURR_VIOLATION
 WHERE VIOLATION_DESCRIPTION <> 'No violations found.'
  AND VIOLATION_DESCRIPTION LIKE '%;'
  AND VIOLATION_DESCRIPTION NOT LIKE '%;%;'
  AND VIOLATION_DESCRIPTION IS NOT NULL
; 

UPDATE CURR_VIOLATION
SET VIOLATION_DESCRIPTION = NULL
WHERE VIOLATION_DESCRIPTION <> 'No violations found.'
  AND VIOLATION_DESCRIPTION LIKE '%;'
  AND VIOLATION_DESCRIPTION NOT LIKE '%;%;'
; 

INSERT INTO TEST_VIOLATION (INSPECTION_ID, VIOLATION_ID, VIOLATION_DESCRIPTION)
SELECT INSPECTION_ID,
 REPLACE(SUBSTR(VIOLATION_DESCRIPTION,
       INSTR(VIOLATION_DESCRIPTION, 'Item ') + 5,
       INSTR(VIOLATION_DESCRIPTION, '-') - (INSTR(VIOLATION_DESCRIPTION, 'Item ') + 5)
 ),' ','') AS violation_id
 , SUBSTR(VIOLATION_DESCRIPTION,
       INSTR(VIOLATION_DESCRIPTION, '-') + 2,
       INSTR(VIOLATION_DESCRIPTION, ';') - INSTR(VIOLATION_DESCRIPTION, '-') - 2
 ) AS violation_description
 FROM CURR_VIOLATION
 WHERE VIOLATION_DESCRIPTION <> 'No violations found.'
  AND VIOLATION_DESCRIPTION IS NOT NULL
; 

UPDATE CURR_VIOLATION
SET VIOLATION_DESCRIPTION = SUBSTR(VIOLATION_DESCRIPTION, INSTR(VIOLATION_DESCRIPTION, ';') + 1)
WHERE VIOLATION_DESCRIPTION LIKE '%;%;' 
 AND VIOLATION_DESCRIPTION <> 'No violations found.'
;
-- END OF ITERATIVE BLOCK

INSERT INTO VIOLATION (VIOLATION_ID, VIOLATION_DESCRIPTION)
SELECT DISTINCT DENSE_RANK() over (order by VIOLATION_ID, VIOLATION_DESCRIPTION),
        VIOLATION_DESCRIPTION
FROM TEST_VIOLATION
 WHERE VIOLATION_ID IS NULL
;

INSERT INTO VIOLATION (VIOLATION_ID, VIOLATION_DESCRIPTION)
SELECT DISTINCT VIOLATION_ID, VIOLATION_DESCRIPTION
FROM TEST_VIOLATION
 WHERE VIOLATION_DESCRIPTION IS NOT NULL
  AND VIOLATION_ID IS NOT NULL
;

-- For VIOLATION_INSPECTION
INSERT INTO VIOLATION_INSPECTION (INSPECTION_ID, VIOLATION_ID)
SELECT DISTINCT tv.INSPECTION_ID, v.VIOLATION_ID
FROM TEST_VIOLATION tv
JOIN VIOLATION v ON v.VIOLATION_DESCRIPTION = tv.VIOLATION_DESCRIPTION
WHERE v.VIOLATION_ID IS NOT NULL
 AND tv.INSPECTION_ID IS NOT NULL
;
```
